```python
#importando bibliotecas
import json
import pymongo
from pymongo import MongoClient
import pprint
import pandas as pd
```

```python
#criando conexao com o banco de dados mongodb
conn = MongoClient('localhost', 27017)
type(conn)
```

```python
# procurando arquivo - insira o caminho onde está salvo .json
with open('C:\\Users\\CRISTINA RAMOS\\Desktop\\Anaconda Projects\\lista_jogadores_chess.json','r') as file:
    arquivo = file.read()
```

```python
#carregar dados do arquivo
json_dados = json.loads(arquivo)
```

```python
# chaves do dicionario
json_dados.keys()
```

```python
#para ver os atributos dentro da key
db = conn.desafiojogadoresdb
type(db)
```

```python
#criando a colection para arquivo
collection = db.dados_pessoais
```

```python
# fazendo insert na base
for dado in json_dados['user']:
    dado_id = collection.insert_one(dado)
    dado_id.inserted_id
```

```python
# contando registros dentro da base
db.dados_pessoais.estimated_document_count()
```

```python
#visualizando dados
for dado in db.dados_pessoais.find():
    pprint.pprint(dado)
```
