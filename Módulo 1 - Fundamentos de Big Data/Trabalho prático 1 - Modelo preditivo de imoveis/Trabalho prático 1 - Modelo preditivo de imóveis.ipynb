### Importando dados

```python
#importar dados
import pandas as pd 
import plotly.express as px
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("custo_diaria.csv", sep=';')
```

### Primeiras análises

```python
df
```

```python
df.describe()
```

```python
df.shape
```

### Criando Variaveis x, y

```python
#criar variavel x (qtd pessoas)
x_df = df.iloc[:,0].values
x_df
```

```python
#criar variavel y (valor_diaria)
y_df = df.iloc[:,1].values
y_df
```

```python
#matriz
np.corrcoef(x_df, y_df)
```

### Criando a matriz de correlação

```python
df.corr()
```

```python
# criar matriz de correlação
plt.figure(figsize=(8,6))
correlation_map = sns.heatmap(df.corr(), annot=True, linewidths=.3, cmap="coolwarm")
```

- Relação forte positiva pois a medida que aumenta a quantidade de pessoas o valor do aluguel aumenta

- 98% do valor cobrado pelas diárias pode ser explicado pela quantidade de pessoas alugando o imóvel

### Tratando dados para o modelo

```python
# visualizar variável
x_df.shape
```

```python
#visualizando formato da variável
x_df
```

```python
#transformando em uma matriz
x_df = x_df.reshape(-1,1)
x_df
```

```python
#matriz com duas colunas pronta
x_df.shape
```

### Aplicando os dados ao Algoritmo de machine de regressão

```python
from sklearn.linear_model import LinearRegression

#cria um objeto do tipo regressão linear
previsao_custo_diaria = LinearRegression()
```

```python
type(previsao_custo_diaria)
```

### Treinamento do modelo

```python
previsao_custo_diaria.fit(x_df, y_df)
```

```python
#verificando a interceptação (constante)
previsao_custo_diaria.intercept_
```

```python
# verificando coeficiente da reta (beta)
previsao_custo_diaria.coef_
```

```python
#fazendo a previsao do valor do imóvel alugado
previsao_precos = previsao_custo_diaria.predict(x_df) 
previsao_precos
```

### Visualizando em gráficos

```python
# transforma a matriz em um vetor
x_df.ravel() 
```

```python
#observando a variavel y
y_df
```

```python
grafico_p=px.scatter(x=x_df.ravel(), y=y_df)
grafico_p.show()
```

```python
#inserindo a reta de regressão linear
grafico_p= px.scatter(x= x_df.ravel(), y= y_df)
grafico_p.add_scatter(x= x_df.ravel(), y = previsao_precos, name = 'Reta de previsões')
grafico_p.show()
```

```python
# ************ verificando os valores de um determinada previsao baseada na quantidade de pessoas ****************
qtd_pessoas = 57
resultado = previsao_custo_diaria.intercept_ + previsao_custo_diaria.coef_ * qtd_pessoas
print("O valor da diária fica aproximadamente:",resultado)
```

```python
#verificando de modo direto (outra forma)
resultado2 = previsao_custo_diaria.predict([[15]])
print("O valor da diária fica aproximadamente:",resultado2)
```

### Verificando as estatisticas do algoritmo

```python
#indica a qualidade do algoritmo (quanto mais proximo de 1, melhor a qualidade do modelo) - 95% 
previsao_custo_diaria.score(x_df, y_df)
```

```python
# visualizando os erros do modelo
from sklearn.metrics import mean_squared_error, mean_absolute_error

MSE = mean_squared_error(y_df, previsao_precos)
MSE
```

```python
# calculando MAE - media dos erros absolutos
mean_absolute_error(y_df, previsao_precos)
```

```python
# calculando o RMSE - tirar a raiz quadrada do MSE
np.sqrt(MSE)
```
